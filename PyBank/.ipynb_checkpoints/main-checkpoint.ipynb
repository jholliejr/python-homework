```python
# Creating a script for analyzing the financial records for my company
# Financial dataset called budget_data.csv will be provided
# Analysis of the data should include the following:
# 1. total number of months included in the dataset
# 2. net total amount of profit/losses over the entire period
# 3. average of the changes in profit/losses over the entire period
# 4. greatest increase in profits (date and month) over the entire period
# 5. greatest decrease in losses (date and month) over the entire period
#
# 
# Must pull in pandas and dataset
# Get a description of the data set
# Sum the net profits/losses
# Determine the cumulative returns/changes between each day's profit/loss, compared to the previous day
# Find the average change/return but dividing by total months
# Identify largest increase
# Identify largest decrease
#
# Pull panda and data set
import pandas as pd
from pathlib import Path
bank_data = Path("budget_data.csv")
#
# Set dataframe to read dataset
bank_df = pd.read_csv(bank_data)
#
# Confirm data was imported and read properly
# bank_df.head()
#
# Begin with description of dataset
# bank_df.info()
# bank_df.describe()
# bank_df['Date'].value_counts()
# bank_df.columns
# blank comment
# print(bank_df['Profit/Losses'].sum()
# blk comm
# blk comm
net_total_months = bank_df['Date'].count()
net_total_amount = bank_df['Profit/Losses'].sum()
#
# Calculating daily differnces/returns
daily_change = (bank_df['Profit/Losses'] - bank_df['Profit/Losses'].shift(1))
# data check
# print(daily_change.sum() / net_total_months)
# Add additional column for differnces
bank_df['Daily Differences'] = daily_change
# check column added
# print(bank_df)

# get the largest profit increase
lgst_pincrease = bank_df.nlargest(1,'Daily Differences')
#
# get the largest profit decrease
lgst_ldecrease = bank_df.nsmallest(1, 'Daily Differences')
# data check
# print(cumulative_change.head())
#
cumulative_average = round(daily_change.sum() / net_total_months, 2)
# data check
# print(cumulative_average)
# blk comm
# greatest_increase = bank_df.nlargest(0, "Profit/Losses")
#
# blk comm
# blk comm
# Output
print("Financial Analysis of the dataset 'budget_data.csv'")
print("~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~*~")
# comment
print(f"Total Months:  {net_total_months}")
print(f"Total Profit/Losses:  ${net_total_amount}")
print(f"Average Change:  ${cumulative_average}")
print("Greatest Increase in Profits:", lgst_pincrease)
print("Greatest Decrease in Profits:", lgst_ldecrease)
# comment
# comment
```
